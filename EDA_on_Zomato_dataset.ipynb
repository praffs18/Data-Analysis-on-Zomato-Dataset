```python
# Now lets load the 2nd csv file
data_country = pd.read_csv('/content/drive/MyDrive/Data Science/EDA Projects/EDA on Zomato Dataset/CountryCode.csv')
data_country
```

```python
# Lets merge this to dataset
data_1 = pd.merge(data,data_country, on = 'Country Code' , how = 'left')
data_1.head()
```

# Insights

```python
# Now lets check how many outlets are on Zomato
data_1['Restaurant Name'].value_counts()
```

There are 7437 outlets(Restaurants) are on Zomato
1. CCD, Dominos, Subway, Green chick chop & McDonalds are the top 5

```python
# Lets see countrywise distribution
country_zomato = pd.DataFrame(data_1['Country'].value_counts())
country_zomato['%'] = round(country_zomato['Country']*100/len(data_1),2)
country_zomato.head()
```

```python
# Lets combine small values
top_2 = country_zomato[0:2]
others = pd.DataFrame(country_zomato[2:].sum())
others = others.T
others = others.rename(index = {0:'others'})
country = top_2.append(others)
country
```

From this we can say that majority of the restaurants are from India & US

```python
# Plot of top 5 country who has most of the restaurants
plt.figure(figsize = (12,6))
explode = [0,0.3,0.3]
plt.pie(country['%'],labels = country.index,autopct='%1.2f%%',explode = explode)
plt.title("Pie chart of use of Zomato", fontsize = 15)
plt.show()
```

```python
# Lets check top 10 cuisine
cuisines = pd.DataFrame(data_1['Cuisines'].value_counts()).head(10)
cuisines = cuisines.reset_index().rename(columns = {'index':'Cuisines', 'Cuisines':'Count'})
cuisines
```

```python
plt.figure(figsize = (13,6))
plots = sns.barplot(data = cuisines , x = 'Cuisines', y = 'Count')
for bar in plots.patches:
  plots.annotate(format(bar.get_height()),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=12, xytext=(0, 8),
                   textcoords='offset points')
plt.title('Top 10 Cuisines', fontsize = 15)
plt.xticks(rotation = 90)
plt.show()
```

```python
# What type of currencies accepted by zomato
currency = pd.DataFrame(data_1.groupby(['Country'])['Currency'].value_counts())
currency = currency.rename(columns = {'Currency':'Count'})
currency
```

1. Four countries are using Dollar($) those are : United States, Singapore, Canada & Australia
2. Rest all are using their own currency for transaction

```python
# How many reastaurants have online delivery services
online_delivery = data_1.groupby('Country')['Has Online delivery'].value_counts()
online_delivery
```

1. Only India and UAE has online delivery services rest are not

```python
# Distribution of price range
plots = sns.countplot(x = data_1['Price range'])
for bar in plots.patches:
  plots.annotate(format(bar.get_height()),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=12, xytext=(0, 8),
                   textcoords='offset points')
plt.title('Countplot of price range')
plt.ylim(0,5000)
plt.show()
```

```python
# Aggregate rating distribution
plt.figure(figsize = (15,6))
sns.countplot(x = data_1['Aggregate rating'])
plt.title('Countplot of Aggregate rating')
plt.show()
```

```python
# Lets get ratings dataframe 
ratings = data_1.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

# Observations
1. When Rating is between 4.5 to 4.9---> Excellent
2. When Rating are between 4.0 to 3.4--->very good
3. when Rating is between 3.5 to 3.9----> good
4. when Rating is between 3.0 to 3.4----> average
5. when Rating is between 2.5 to 2.9----> average
6. when Rating is between 2.0 to 2.4----> Poor

```python
# There are lot of Not rated restaurants 
data_1[data_1['Rating text'] == 'Not rated'].groupby(['Country']).size()
```

So, Most of the Restaurants are from India who does not have rating

```python
## Count plot
plt.figure(figsize = (15,6))
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
plt.show()
```
